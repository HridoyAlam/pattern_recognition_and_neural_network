```python
import numpy as np
import pandas as pd
import re
```

```python
df_livedata = pd.read_csv('livedata.csv')
df_teamrank = pd.read_csv('clean_team_rank.csv')
```

```python
df_livedata.head(10)
```

```python
df_teamrank.head(15)
```

```python
results_df = df_livedata.copy()
rank_df = df_teamrank.copy()
```

```python
combined_df = pd.merge(results_df, rank_df.rename(columns={'team_name': 'team_won', 'team_rank': 'winning_rank'}), on='team_won')
```

```python
combined_df = pd.merge(combined_df , rank_df.rename(columns={'team_name': 'team_lost', 'team_rank': 'losing_rank'}), on='team_lost')
```

```python
combined_df.head()
```

```python
combined_df.columns
```

```python
df = combined_df[['team_won','team_lost', 'winning_score', 'losing_score', 'winning_rank', 'losing_rank']]
df.head(25)
```

```python
df.to_csv('live_rank_data.csv')
```
