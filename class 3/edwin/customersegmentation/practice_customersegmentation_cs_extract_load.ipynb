```python
def splitname(row):
    
    row['StockCode'] = row['InvoiceNo'].split(';')[1]
    row['Description'] = row['InvoiceNo'].split(';')[2]
    row['Quantity'] = row['InvoiceNo'].split(';')[3]
    row['InvoiceDate']=row['InvoiceNo'].split(';')[4]
    row['UnitPrice']=row['InvoiceNo'].split(';')[5]
    row['CustomerID']=row['InvoiceNo'].split(';')[6]
    row['Country']=row['InvoiceNo'].split(';')[7]
    row['InvoiceNo'] = row['InvoiceNo'].split(';')[0]
    return row
```
